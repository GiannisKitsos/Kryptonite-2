```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

x = np.load('Datasets/kryptonite-14-X.npy')
y = np.load('Datasets/kryptonite-14-y.npy')
scaler = StandardScaler()   
x_std = scaler.fit_transform(x)  # z-normalisation of features
  
pca = PCA(n_components=2)  # PCA object to compute the first 2 pc_s
z = pca.fit_transform(x_std) # fits the model to the standardised data

colours = ['red' if label==0 else 'blue' for label in y]

plt.scatter(z[:,0], z[:,1], c=colours, alpha=0.4, edgecolors='k')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('PCA projection for n=10 dataset. Red=0, Blue=1')
plt.show()
```
